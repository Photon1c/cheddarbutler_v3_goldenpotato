```python
import alpaca_trade_api as tradeapi
import config
import logging
from alpaca_trade_api.rest import REST, TimeFrame
import datetime as dt
import pandas as pd
from datetime import date


#ally api endpoint creation
api = tradeapi.REST(key_id=config.LIVE_API_KEY, secret_key=config.LIVE_SECRET_KEY, api_version='v2',
                                        base_url=config.LIVE_API_BASE_URL)

source_greeting = str("For the data source: Enter M if you will be running Golden Potato using a list of default \ndaily top moving stocks. Enter L if you will be manually entering a list of stock tickers. \nFinally, enter F if you want to enter the name of a CSV file containing a list of stock tickers. \nPlease refer to the documentation for the preferred format. ")



#Collect inputs from user

quick_list = []

def input_collector():
    print(source_greeting)
    source = input("Enter M, L, or F: ")
    print(source)
    date_set = input("Default 2 week historical data or enter customized date range? Enter D for default or C to customize: ")
    s_date_list = []
    e_date_list = []
    if source == "L":
        stock_list = input("Enter stock ticker: ")
    for i in stock_list:
        templ = stock_list
        quick_list.append(templ)
        out_quick = pd.DataFrame(quick_list)
        out_quick.to_csv(r'D:\mainline\goldenpotato\outputs\quick_list.csv')
    if date_set == "C":
        s_date = input("Enter start date in YYYY-MM-DD format: ")
        for i in s_date:
            temps = s_date
            s_date_list.append(temps)
        e_date = input("Enter end date in similar format: ")
        for i in e_date:
            tempe = e_date
            e_date_list.append(tempe)
    print(quick_list)
    print(s_date)
    print(e_date)
    format = "%Y-%m-%d"

    try:
        datetime.datetime.strptime(s_date, format)
        datetime.datetime.strptime(e_date, format)
        print("This is the correct date string format.")
        print("Thank you. Adding to Golden Potato.")
    except ValueError:
        print("This is the incorrect date string format. It should be YYYY-MM-DD")

    
    
    
    
    
    

input_collector()



  



















excel_file = pd.read_csv(r"C:\Users\Spark\OneDrive\Desktop\currentprojects\tradesetandforget\csvs\q2.csv")

now = date.today()
today = now.strftime('%Y-%m-%d')
yesterday = (now - pd.Timedelta('1day')).strftime('%Y-%m-%d')
ninedaysago = (now - pd.Timedelta('9day')).strftime('%Y-%m-%d')


Symbols = excel_file['symbol'].tolist()
def start():
    sym = []
    dif = []
    result_df = pd.DataFrame()
    for i in Symbols:
        # process quote
        quote_iter = api.get_bars(i, TimeFrame.Day, start = ninedaysago, end = yesterday, limit=9)._raw
        # See how much stock moved in that timeframe.
        quote_df = pd.DataFrame(quote_iter)
        ninedago = quote_df.iloc[0]['c']
        pricenow = quote_df.iloc[5]['c']
        percent_change = (pricenow - ninedago) / ninedago * 100
        print(i, ' moved {:.2f}% over the last 9 days'.format(percent_change))
        sym.append(str(i))
        dif.append(percent_change)
    result_df["Symbol"] = sym
    result_df["Difference"] = dif
    result_df["Difference"] = result_df["Difference"].map('{:.2f}%'.format)
    result_df.to_csv("successfullscreen03172022deltastock")
    print(result_df)

start()


```
